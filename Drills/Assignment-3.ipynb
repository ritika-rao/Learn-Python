## Question

Design an **interactive program** to **connect to alumni** of your university. 

Get details of alumni (name, email, phone, company name, design, year_passed_out) from alumni. If the alumni is from **batch before 2012** then check if he would be interested. If **he enters yes**, then asks for getting associated with the University for which of the following 

a) Industry mentor 

b) Projects 

c) Guest Lecture 

d) Placements. 

In each of the following get the date and time suitable to mentor.

If the alumni is from **batch after 2012** then asks the involvement for 

a) Workshops 

b) Internship, 

accordingly take the date and time. 

If **interested in internship** then ask from alumni, enter number of students and skill set required for internship. The number of alumni can be changing every time program is run.

**Note: -** Take **atleast 5 alumni**. Display all the parameters in clear and good format.  Use functions as deemed appropriate.

```python
class Alumni:
    def __init__(self, name, email, phone, company_name, year_passed_out):
        self.name=name
        self.email=email
        self.phone=phone
        self.company_name=company_name
        self.year_passed_out=year_passed_out
        self.associate=None
        self.DateNtime=None
        self.no_ofSt=None
        self.skill=None
    def setData(self,associate,DateNtime):
        self.associate=associate
        self.DateNtime=DateNtime
        if associate=="Internship":
            no_ofSt=input("enter number of students")
            skill=input("enter skill set required for internship")
            self.no_ofSt=no_ofSt
            self.skill=skill
    def showData(self):
        print("name is ",end="")
        print(self.name)
        print("email is ",end="")
        print(self.email)
        print("phone is ",end="")
        print(self.phone)
        print("company name is ",end="")
        print(self.company_name)
        print("year passed out is ",end="")
        print(self.year_passed_out)
        print("associate is ",end="")
        print(self.associate)
        if self.associate=="Internship":
            print("no.ofSt is ",end="")
            print(self.no_ofSt)
            print("skill is ",end="")
            print(self.skill)
        print("DateNtime is ",end="")
        print(self.DateNtime)
        print('-'*60)
#main
alumnis=[]
print("press yes if u want to enter details of alumni")
choice=input()
while choice=="yes" or choice=="YES":
    name=input("enter name")
    email=input("enter email")
    phone=int(input("enter phone"))
    company_name=input("enter company name")
    year_passed_out=int(input("enter year passed out"))
    alumni=Alumni(name, email, phone, company_name, year_passed_out)
    interest=input("type yes if interested")
    if interest=="yes" or interest=="YES":
        if alumni.year_passed_out<2012:
            associate=input("a) Industry mentor    b) Projects     c) Guest Lecture     d) Placements")
            date=input("enter date and time suitable")
            alumni.setData(associate,date)
            
        elif alumni.year_passed_out>2012:
            associate=input("a) Workshops     b) Internship")
            date=input("enter date and time suitable")
            alumni.setData(associate,date)
    else:
        alumni.setData("none","none")
    alumnis.append(alumni)
    print("press yes if u want to enter details of alumni")
    choice=input()
print()
print("-------------SHOWING DATA-----------")
for a in alumnis:
    a.showData()
```
